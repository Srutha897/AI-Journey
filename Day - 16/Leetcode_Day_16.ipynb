Leetcode #435 - Non-overlapping Intervals

```python
class Solution:
    def eraseOverlapIntervals(self, intervals: List[List[int]]) -> int:
        if not intervals:
            return 0

        # 1. Sort by end time
        intervals.sort(key=lambda x: x[1])

        count = 0
        # Initialize 'end' with the end time of the first interval
        current_end = intervals[0][1]

        # 2. Iterate starting from the second interval
        for i in range(1, len(intervals)):
            start_i, end_i = intervals[i]

            if start_i < current_end:
                # Overlap detected! We "remove" this interval.
                count += 1
            else:
                # No overlap. Update the end marker to the current interval's end.
                current_end = end_i

        return count
```
