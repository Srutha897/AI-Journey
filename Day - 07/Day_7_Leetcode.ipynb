Leetcode #424 - Longest Repeating Character Replacement

```python
class Solution:
    def characterReplacement(self, s: str, k: int) -> int:
        left = 0
        max_length = 0
        max_freq = 0
        count = {}

        for right in range(len(s)):
            # 1. Add current character to frequency map
            count[s[right]] = count.get(s[right], 0) + 1

            # 2. Track the most frequent character in the current window
            max_freq = max(max_freq, count[s[right]])

            # 3. Check if window is valid
            # (Window size) - (count of most frequent char) = chars to replace
            while (right - left + 1) - max_freq > k:
                count[s[left]] -= 1
                left += 1
                # because we only care when max_freq increases to find a NEW max_length.

            # 4. Update max length with the window size
            max_length = max(max_length, right - left + 1)

        return max_length
```

Leetcode #3 - Longest Substring Without Repeating Characters

```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        left = 0
        max_length = 0
        window = set()
        for right in range(len(s)):
            while(s[right] in window):
                window.remove(s[left])
                left+=1
            window.add(s[right])
            max_length=max(max_length,len(window))
        return max_length
```
