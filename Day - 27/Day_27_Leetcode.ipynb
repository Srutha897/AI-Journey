Leetcode #2325 - Decode the Message

```python
class Solution:
    def decodeMessage(self, key: str, message: str) -> str:
        dict={}
        temp=97
        secret_message=""
        for i in key:
            if i!=" " and i not in dict:
                dict[i]=chr(temp)
                temp+=1
        for i in message:
            if i==" ":
                secret_message+=" "
            else:
                secret_message+=dict[i]
        return secret_message
```

Leetcode #771 - Jewels and Stones

```python
class Solution:
    def numJewelsInStones(self, jewels: str, stones: str) -> int:
        dict={}
        count=0
        for i in stones:
            if i in jewels:
                if i not in dict:
                    dict[i]=1
                    count+=1
                else:
                    dict[i]+=1
                    count+=1
            else:
                count+=0
        return count
```
